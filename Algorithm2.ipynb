```python
import unittest
import time
from math import floor

from math import ceil
import matplotlib.pyplot as plt

def Algorithm2(A):
    Compute(A)
    Dump(A)
    return

def Compute(A):
    N = len(A)
    size = floor(N ** 0.5) + 1
    Z = [[] for i in range(size)]

    # Populate Z with elements from A
    for i in range(N):
        Z[A[i] % size].append(A[i])

    counter = 1
    for i in range(size):
        while Z[i]:
            A[counter - 1] = Z[i].pop(0)
            counter += 1

    for i in range(N):
        Z[A[i] // size].append(A[i])

    counter = 1
    for i in range(size):
        while Z[i]:
            A[counter - 1] = Z[i].pop(0)
            counter += 1

    return

def Dump(B):
    for i in range(len(B)):
        print(B[i], end="")
    print(" ")
    return


class TestAlgorithm2(unittest.TestCase):
    
    def read_array_from_file(self, filename):
        with open(filename, 'r') as file:
            lines = file.readlines()
            return [int(num) for line in lines for num in line.split()]

    def measure_runtime(self, A):
        start_time = time.time()
        Algorithm2(A)
        end_time = time.time()
        return end_time - start_time

    def compute_statistics(self, runtimes):
        min_time = min(runtimes)
        max_time = max(runtimes)
        avg_time = sum(runtimes) / len(runtimes)
        return min_time, avg_time, max_time

    def test_duplicate_array(self):
        # test_cases = ['duplicate_array_10.txt', 'duplicate_array_20.txt', 'duplicate_array_30.txt', 'duplicate_array_40.txt', 'duplicate_array_50.txt']
        test_cases = ['duplicate_array_3.txt', 'duplicate_array_8.txt', 'duplicate_array_10.txt', 'duplicate_array_13.txt', 'duplicate_array_15.txt']
        runtimes = []

        for filename in test_cases:
            A = self.read_array_from_file(filename)
            runtime = self.measure_runtime(A)
            runtimes.append(runtime)
            print(f"\n\nRuntime for duplicate array (N = {len(A)}): {runtime} seconds")

        min_time, avg_time, max_time = self.compute_statistics(runtimes)
        print(f"\nBest time: {min_time} seconds")
        print(f"Average time: {avg_time} seconds")
        print(f"Worst time: {max_time} seconds\n")

    def test_reverse_sorted_array(self):
        print("\nRunning tests on reverse-sorted array...")
        test_cases = ['reverse_sorted_array_3.txt', 'reverse_sorted_array_8.txt', 'reverse_sorted_array_10.txt', 'reverse_sorted_array_13.txt', 'reverse_sorted_array_15.txt']
        runtimes = []

        for filename in test_cases:
            A = self.read_array_from_file(filename)
            runtime = self.measure_runtime(A)
            runtimes.append(runtime)
            print(f"\n\nRuntime for reverse_sorted_array (N = {len(A)}): {runtime} seconds")

        min_time, avg_time, max_time = self.compute_statistics(runtimes)
        print(f"\nBest time: {min_time} seconds")
        print(f"Average time: {avg_time} seconds")
        print(f"Worst time: {max_time} seconds\n")

    def test_sorted_array(self):
        print("\nRunning tests on sorted array...")
        test_cases = ['sorted_array_3.txt', 'sorted_array_8.txt', 'sorted_array_10.txt', 'sorted_array_13.txt', 'sorted_array_15.txt']
        runtimes = []

        for filename in test_cases:
            A = self.read_array_from_file(filename)
            runtime = self.measure_runtime(A)
            runtimes.append(runtime)
            print(f"\n\nRuntime for sorted_array (N = {len(A)}): {runtime} seconds")

        min_time, avg_time, max_time = self.compute_statistics(runtimes)
        print(f"\nBest time: {min_time} seconds")
        print(f"Average time: {avg_time} seconds")
        print(f"Worst time: {max_time} seconds\n")

    def test_random_array(self):
        print("\nRunning tests on random array...")
        test_cases = ['random_array_3.txt', 'random_array_8.txt', 'random_array_10.txt', 'random_array_13.txt', 'random_array_15.txt']
        runtimes = []

        for filename in test_cases:
            A = self.read_array_from_file(filename)
            runtime = self.measure_runtime(A)
            runtimes.append(runtime)
            print(f"\n\nRuntime for random_array (N = {len(A)}): {runtime} seconds")

        min_time, avg_time, max_time = self.compute_statistics(runtimes)
        print(f"\nBest time: {min_time} seconds")
        print(f"Average time: {avg_time} seconds")
        print(f"Worst time: {max_time} seconds\n")

def measure_runtime(A):
        start_time = time.time()
        Algorithm2(A)
        end_time = time.time()
        return end_time - start_time

def calculate_predicted_runtime(N):
    return N

if __name__ == '__main__':
    unittest.main(argv=[''], exit=False)

    test_sizes = [3, 8, 10, 13, 15, 20, 25, 40, 75, 100, 250, 400, 475, 500]
    # Best case scenario: array is already sorted
    # Different array sizes for best case scenario
    best_case_measured_times = []
    for size in test_sizes:
        print(f"Running tests on best case scenario with array size {size}...")
        A = list(range(1, size+1))
        print(f"Array: {A}")
        print("SortedArray: ", end="")
        best_case_measured_times.append(measure_runtime(A))
        print("\n")
    
    # Get predicted runtimes for best case scenario
    best_case_predicted_times = [calculate_predicted_runtime(size) for size in test_sizes]

    # Plot a line graph predicted runtimes for best case scenario
    plt.plot(test_sizes, best_case_predicted_times)
    plt.xlabel('Array Size')
    plt.ylabel('Time (s)')
    plt.title('Predicted Times for Best Case Scenario')
    plt.show()

    # plot a line graph of measured runtimes for best case scenario
    plt.plot(test_sizes, best_case_measured_times, c='r')
    plt.xlabel('Array Size')
    plt.ylabel('Time (s)')
    plt.title('Measured Times for Best Case Scenario')
    plt.show()
    

    # Worst case scenario: array is reverse sorted
    # Different array sizes for worst case scenario
    worst_case_measured_times = []
    for size in test_sizes:
        print(f"Running tests on worst case scenario with array size {size}...")
        A = list(range(size, 0, -1))
        print(f"Array: {A}")
        print("ReverseSortedArray: ", end="")
        worst_case_measured_times.append(measure_runtime(A))
        print("\n")

    # Get predicted runtimes for worst case scenario
    worst_case_predicted_times = [calculate_predicted_runtime(size) for size in test_sizes]

    # Plot a line graph predicted runtimes for worst case scenario
    plt.plot(test_sizes, worst_case_predicted_times)
    plt.xlabel('Array Size')
    plt.ylabel('Time (s)')
    plt.title('Predicted Times for Worst Case Scenario')
    plt.show()

    # plot a line graph of measured runtimes for worst case scenario
    plt.plot(test_sizes, worst_case_measured_times, c='r')
    plt.xlabel('Array Size')
    plt.ylabel('Time (s)')
    plt.title('Measured Times for Worst Case Scenario')
    plt.show()

```
